```python
import matplotlib.pyplot as plt

# y = -5x+17 ;  y = -x+5 
x1=[]
y1 =[]
for i in range(-50,50):
    x1.append(i)
    y1.append((-5*i)+17)
    

plt.plot(x1, y1, label = "line 1")

x2=[]
y2= []
for i in range(-50,50):
    x2.append(i)
    y2.append((-i)+5)
    



plt.plot(x2, y2, label = "line 2")


plt.xlabel('x - axis')
plt.ylabel('y - axis')


plt.title('Two or more lines on same plot with suitable legends ')

plt.legend()

plt.show()
```
